7. Write a python program to find the longest words.


```python
l='my name is bapu kkkkk shsshhs  xbxghx'
#split into list of words
k = l.split()
print(k)
#find max length of word
print(max(k))
print(len(max(k)))
```

```python
def find_max_length_word(file1):
  with open(file1, 'r') as f:

    #file converting into list of words
    word = f.read().split()
    print(word)

    #find max length of word
    max_length = max(len(i) for i in word)

    # printing max word
    longest_word = [i for i in word if len(i)==max_length]
    print('longest word = ',longest_word)
    print('maximum length of word = ',max_length)



find_max_length_word('sample1.txt')
```

13. Write a Python program to read a random line from a file.
