Create a Python program that allows a user to input student names along with their marks and then calculates useful statistics.
Your program should:
1. Let the user input multiple students with their marks (name + integer score).
2. After input is complete, display:
   - Average marks
   - Highest marks and student(s) who scored it
   - Lowest marks and student(s) who scored it
   - Total number of students

Bonus:
- Allow the user to enter all data first, then view the report
- Format output clearly in a report-style layout
- Prevent duplicate student names


```python
class Student:
    def __init__(self):
        self.students = {}


    def add_student(self):
        while True:
            name = input("Enter student name (or type '0' to finish): ").strip()

            if name.lower() == '0':
                break

            if name in self.students:
                print(" Student already exists. Duplicate names are not allowed.\n")
                continue

            try:
                marks = int(input(f"Enter marks for {name}: "))
                self.students[name] = marks
                print(" Student added successfully!\n")
            except ValueError:
                print(" Please enter valid integer marks.\n")

    def generate_report(self):
        if not self.students:
            print("No student data available.")
            return

        marks_list = list(self.students.values())
        avg_marks = sum(marks_list) / len(marks_list)

        max_marks = max(marks_list)
        min_marks = min(marks_list)

        top_students = [name for name, marks in self.students.items() if marks == max_marks]
        low_students = [name for name, marks in self.students.items() if marks == min_marks]

        print("\n" + "-" * 40)
        print("STUDENT REPORT")
        print("-" * 40)
        print(f"Total Students   : {len(self.students)}")
        print(f"Average Marks    : {avg_marks:.2f}")
        print(f"Highest Marks    : {max_marks}")
        print(f"Topper(s)        : {', '.join(top_students)}")
        print(f"Lowest Marks     : {min_marks}")
        print(f"Lowest Scorer(s) : {', '.join(low_students)}")
        print("=" * 40)


```

```python

manager = Student()

print("=== Student Marks Entry System ===")
manager.add_student()

choice = input("\nDo you want to view the report? (yes/no): ").lower()
if choice == 'yes':
    manager.generate_report()
else:
    print("Report generation skipped.")
```
