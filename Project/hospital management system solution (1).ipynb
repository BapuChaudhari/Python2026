```python
#patient information
class Patient:
  def __init__(self,name,status):
    self.name = name
    self.status = status #0=normal 1=urgent 2= super-urgent
  def __str__(self):
    status = ['normal','urgent','super-urgent']
    return "{} ({})".format(self.name,status[self.status])

# creation of specialisation class
class Specialisation:
  def __init__(self,name,max_capacity=15):
    self.name = name
    self.queue = []
    self.max_capacity = max_capacity

  def is_full(self):
    return len(self.queue) >= self.max_capacity

  def add_patient(self,patient):
    if self.is_full():
      print("queue is full")
      return

    #add patient by status wise urgent gives priority
    if patient.status == 2:
      self.queue.insert(0,patient)

    elif patient.status == 1:
      index = 0
      while index < len(self.queue) and self.queue[index].status == 2:
        index +=1
      self.queue.insert(index,patient)

    else:
      self.queue.append(patient)

  def get_next_patient(self):
    if not self.queue:
      print("no patient in queue")
      return None
    return self.queue.pop(0)

  #remove patient
  def remove_patient(self,name):
    for patient in self.queue:
      if patient.name == name:
        self.queue.remove(patient)
        print("{} removed successifully".format(name))
        return
    print('patient not found')

  def list_patients(self):
    if not self.queue:
      print("no patient in this specialisation")
    else:
      for patient in self.queue:
        print(patient)

class OperationManager:
  def __init__(self):
    self.specialisation = {
        1: Specialisation('general'),
        2: Specialisation('cardiology'),
        3: Specialisation('neurology')
    }

  def run(self):
    while True:
      print("\n1. add patient")
      print("2. list patients")
      print("3. get next patient")
      print("4. remove patient")
      print("5. exit")

      choice = int(input("enter the choice = "))

      if choice == 1:
        spec_id = int( input("enter specialisation id 1. general 2. cardiology 3. neurology = "))
        name = input("enter the name of patient = ")
        status = int(input('enter status 0.normal 1.urgent 2.super-urgent = '))
        patient = Patient(name,status)
        self.specialisation[spec_id].add_patient(patient)
      elif choice == 2:
        spec_id = int( input("enter specialisation id 1. general 2. cardiology 3. neurology = "))
        self.specialisation[spec_id].list_patients()
      elif choice == 3:
        spec_id = int( input("enter specialisation id 1. general 2. cardiology 3. neurology = "))
        patient = self.specialisation[spec_id].get_next_patient()
        if patient:
          print("next patient = ",patient)

      elif choice == 4:
        spec_id = int( input("enter specialisation id 1. general 2. cardiology 3. neurology = "))
        name = input("enter the name of patient = ")
        self.specialisation[spec_id].remove_patient(name)
      elif choice == 5:
        print("existed from system")
        break
      else:
        print("invalid choice")













```

```python
m=OperationManager()
m.run()
```
