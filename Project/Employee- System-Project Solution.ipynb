Employees System Project

The Employees System Project is a simple Python-based project that showcases the use of object-oriented programming (OOP) principles.

It includes three main classes: Employee, EmployeesManager, and FrontendManager. This project provides a basic example of managing employee data and interactions using OOP concepts.

Introduction

The Employees System Project demonstrates the implementation of object-oriented programming concepts in Python. It encompasses three primary classes, each serving a distinct purpose:

Employee
The Employee class represents an individual employee with the following attributes:

•	name: The name of the employee.

•	age: The age of the employee.

•	salary: The salary of the employee.

This class provides methods for string representation and formatted output of employee information.

EmployeesManager

The EmployeesManager class is responsible for managing a list of employees. It offers functionalities to:

•	Add a new employee to the list.

•	List all existing employees.

•	Delete employees within a specified age range.

•	Find an employee by their name.

•	Update an employee's salary by name.

FrontendManager

The FrontendManager class provides a user interface for interacting with the EmployeesManager.

 Users can perform actions such as:

•	Adding new employees.

•	Listing existing employees.

•	Deleting employees based on age range.

•	Updating employee salaries by name.


```python
#employee class
class Employee():
  def __init__(self,name,age,salary):
    self.name = name
    self.age = age
    self.salary = salary
  def __str__(self):
    return "Employee Name: {}\nEmployee Age: {}\nEmployee Salary: {}".format(self.name,self.age,self.salary)

class EmployeesManager():
  def __init__(self):
    self.employees = []

  def add_employee(self,employee):
    self.employees.append(employee)

  def list_employees(self):
    for employee in self.employees:
      print(employee)

  def delete_employees(self,age):
    for employee in self.employees:
      if employee.age >= age:
        self.employees.remove(employee)

  def find_employee(self,name):
    for employee in self.employees:
      if employee.name == name:
        return employee

  def update_employee(self,name,salary):
    for employee in self.employees:
      if employee.name == name:
        employee.salary = salary
        return True
    return False

  def __str__(self):
    return "EmployeesManager"

class FrontendManager():
  def __init__(self):
    self.employees_manager = EmployeesManager()

    print("""
    Welcome to the Employees Manager!
    Please select an option:
    1. Adding new employees.
    2. Listing existing employees.
    3. Deleting employees based on age range.
    4. Updating employee salaries by name.
    5. enter for exit
    """)

    while True:
      option = input("Enter your option: ")
      if option == "1":
        name = input("Enter employee name: ")
        age = int(input("Enter employee age: "))
        salary = int(input("Enter employee salary: "))
        employee = Employee(name,age,salary)

        self.employees_manager.add_employee(employee)
        print("Employee added successfully!")

      elif option == "2":
        print("List of employees:")
        self.employees_manager.list_employees()

      elif option == "3":
        age = int(input("Enter age range to delete employees: "))
        self.employees_manager.delete_employees(age)
        print("Employees deleted successfully!")

      elif option == "4":
        name = input("Enter employee name to update salary: ")
        salary = int(input("Enter new salary: "))
        if self.employees_manager.update_employee(name,salary):
          print("Employee salary updated successfully!")
        else:
          print("Employee not found!")

      elif option == "5":
        print("Thank you for using the Employees Manager!")
        break




```
